```python
# Загрузка необходимых модулей
import requests
import os
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
from time import sleep
from bs4 import BeautifulSoup
```

```python
def load_page(url):
    '''
    Функция загрузки веб страницы
    '''
    count = 0
    while True:
        try:
            with requests.Session() as session:
                session.headers['Connection'] = 'keep-alive'
                session.headers['Cache-Control'] = 'max-age=0'
                session.headers['Upgrade-Insecure-Requests'] = '1'
                session.headers['User-Agent'] = 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.116 Safari/537.36 OPR/40.0.2308.81'
                session.headers['Accept'] = 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8'
                session.headers['DNT'] = '1'
                session.headers['Accept-Encoding'] = 'gzip, deflate, lzma, sdch'
                session.headers['Accept-Language'] = 'ru-RU,ru;q=0.8,en-US;q=0.6,en;q=0.4'
                retry = Retry(connect=3, backoff_factor=0.5)
                adapter = HTTPAdapter(max_retries=retry)
                session.mount('http://', adapter)
                session.mount('https://', adapter)
                session.proxies = {
                    # 'http': '185.198.189.21:8080',
                    # 'https': '185.198.189.21:8080'
                }
                response = session.get(url)
            if response.status_code != 200:
                # log.info("Ошибка, код ответа: %s", response.status)
                print('Ошибка, код ответа: %s', response.status_code)
                sleep(1)
                count += 1
                if count > 5:
                    return (-1)
                # Попробуем снова на следующей итерации цикла
                continue
            # Если дошли до сюда, значит ошибок не было
            return response
        except ConnectionError:
            # Выводим свое сообщение плюс стек трассы
            # log.exception("Ошибка ConnectionError")
            print('Ошибка ConnectionError')
            sleep(1)


def download_image(url, class_id, file_name):
    directory = f'images/{class_id}'
    if not os.path.exists(directory):
        os.makedirs(directory)
#     file_name = 0
    while os.path.exists(f'{directory}/{file_name}.jpg'):
        file_name = file_name + 1
    # open in binary mode
    with open(f'{directory}/{file_name}.jpg', "wb") as file:
        # get request
        response = requests.get(url)
        # write to file
        file.write(response.content)
```

```python
classes = {
    0: ['https://auto.ru/cars/vaz/2170/all/?page=', 99],
    1: ['https://auto.ru/cars/ford/focus/all/?page=', 99],
    2: ['https://auto.ru/cars/vaz/2114/all/?page=', 96],
    3: ['https://auto.ru/cars/vaz/2110/all/?page=', 69],
    4: ['https://auto.ru/cars/vaz/2107/all/?page=', 72],
    5: ['https://auto.ru/cars/vaz/niva/all/?page=', 19],
    6: ['https://auto.ru/cars/vaz/kalina/all/?page=', 94],
    7: ['https://auto.ru/cars/vaz/2109/all/?page=', 36],
    8: ['https://auto.ru/cars/volkswagen/passat/all/?page=', 77],
    9: ['https://auto.ru/cars/vaz/21099/all/?page=', 31]
}
```

```python
for class_id in classes:
    file_name = 0
    print(f'Class - {class_id}')
    for page in range(1, classes[class_id][1]+1):
        print(f'Page - {page}/{classes[class_id][1]}')
        url = f'{classes[class_id][0]}{page}'
        response = load_page(url)
        if response == -1:
            continue
        soup = BeautifulSoup(response.text, 'html.parser')
        links = soup.find_all('a', class_='Link ListingItemTitle__link')
        for link in links:
            print(link.get('href'))
            response = load_page(link.get('href'))
            if response == -1:
                continue
            soup = BeautifulSoup(response.text, 'html.parser')
            images = soup.find_all('img', class_='ImageGalleryDesktop__image')
            for image in images:
                # print(f"http:{image.get('src')}")
                download_image(f"http:{image.get('src')}", class_id, file_name)
                file_name += 1
        sleep(5)
```
